```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
print(type(fig))
fig.suptitle("Empty figure")
plt.show()
```

```python
fig = plt.figure()
fig.suptitle("Empty figure")

ax = fig.add_subplot()
ax.set_title("Empty plot")
plt.show()
```

```python
fig = plt.figure()
fig.suptitle("Empty figure")

ax1 = fig.add_subplot(2,2,1)
ax1.set_title("First Location", fontdict= {'fontsize' : 8}, pad=0.0)

ax2 = fig.add_subplot(2,2,2)
ax2.set_title("Second Location", fontdict = {'fontsize' : 8}, pad=0.0)

ax3 = fig.add_subplot(2,2,3)
ax3.set_title("Third Location", fontdict = {'fontsize' : 8}, pad=0.0)

ax4 = fig.add_subplot(2,2,4)
ax4.set_title("Fourth Location", fontdict = {'fontsize' : 8}, pad=0.0)
    

plt.show()
```

```python
x = [1,2,3,4,5]
y = x

# create an empty figure
fig = plt.figure()

# get the axes instance a 1st location in 1x1 grid
ax = fig.add_subplot()

# generate the plot
ax.plot(x, y)

ax.set_title("Simple XY Plot")
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```

```python
def list_square(a_list):
    return [ element ** 2 for element in a_list]
```

```python
x = [1,2,3,4,5]
y = x
z = list_square(x)

print(z)
```

```python
# create an empty figure
fig = plt.figure()

# get the axes instance a 1st location in 1x1 grid
ax = fig.add_subplot()

# generate the plot
ax.plot(x, y, label='y')
ax.plot(x,z, label='z')

ax.set_title("Simple XY Plot")
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend()

plt.show()
```

```python
# create an empty figure
fig = plt.figure(figsize=(10, 6))

# get the axes instance a 1st location in 1x1 grid
ax1 = fig.add_subplot(1,2,1)

# generate the plot y=x
ax1.plot(x, y, label='y')

ax1.set_title("Simple y = x Plot")
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.legend()

# get the axes instance a 1st location in 1x1 grid
ax2 = fig.add_subplot(1,2,2)

# generate the plot y=x
ax2.plot(x,z, label='z')

ax2.set_title("Simple z =  x ** 2 Plot")
ax2.set_xlabel('X')
ax2.set_ylabel('Z')
ax2.legend()

plt.suptitle('Two Plots in a Figure')
plt.show()
```
